```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df= pd.read_excel("/content/Module 9 . Visualization data.xlsx")
```

```python
df
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
stock_columns = ['Zorto', 'laxn', 'Survelia', 'Jecco']
```

```python
stock_columns
```

```python
for col in stock_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df_cleaned = df.dropna(subset=stock_columns)
```

```python
correlation_matrix = df_cleaned[stock_columns].corr()
```

```python
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Stock price Correlation_Matrix", fontdict={'fontsize': 16})
plt.show()
```
