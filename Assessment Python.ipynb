5. Get a list or array of all the states in the data set.

6. What are the top 5 states with the most failed banks?

```python
state_counts = banks["ST"].value_counts()
top_5_states = state_counts.head(5)
print(top_5_states)
```

7. What are the top 5 acquiring institutions?

```python
acquirer_counts = banks['Acquiring Institution'].value_counts()
top_5_acquirers = acquirer_counts.head(5)
print(top_5_acquirers)
```

8. How many banks has the State Bank of Texas acquired? How many of 
them were actually in Texas?

```python
bank_acquired = banks[banks['Acquiring Institution'] == 'State Bank of Texas']
total_acquired = bank_acquired.shape[0]
texas_acquired = bank_acquired[bank_acquired['ST'] == 'TX'].shape[0]
print(f"Total Banks acquired by state bank of Texas: {total_acquired}")
print(f"Banks acquired in Texas: {texas_acquired}")
```

9. What is the most common city in California for a bank to fail in?

```python
california_banks = banks[banks['ST'] == 'CA']
city_counts = california_banks['City'].value_counts()
most_common_city = city_counts.idxmax()
count = city_counts.max()
print(f"The most common city in California for a bank to fail in is: {most_common_city}({count}failures)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("D:/TOPS TECHNO/PYTHON/Assessment/historical_automobile_sales.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
yearly_sales = df.groupby('Year')['Automobile_Sales'].sum().reset_index()
plt.figure(figsize = (12,6))
plt.plot(yearly_sales['Year'],yearly_sales['Automobile_Sales'],color = 'green',linewidth = 2)
plt.title("Automobile Sales by Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Automobile Sales", fontsize=12)
plt.grid(True)
plt.xticks(yearly_sales['Year'])
plt.tight_layout()
```

2. Plot different lines for categories of vehicle type and analyze the trend 
to answer the question Is there a noticeable difference in sales trends 
between different vehicle types during recession periods?

```python
recession_data = df[df['Recession'] == 1]
grouped = recession_data.groupby(['Year','Vehicle_Type'])['Automobile_Sales'].sum().reset_index()
plt.figure(figsize = (12,6))
sns.lineplot(data = grouped, x = 'Year', y = 'Automobile_Sales')
plt.title("Automobile Sales Trends by Vehicle Type During Recession Years", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.grid(True)
plt.tight_layout()
```

3. Use the functionality of Seaborn Library to create a visualization to compare 
the sales trend per vehicle type for a recession period with a non- recession 
period.

```python
grouped = df.groupby(['Year', 'Recession', 'Vehicle_Type'])['Automobile_Sales'].sum().reset_index()
plt.figure(figsize=(14, 7))
sns.lineplot(data=grouped, x='Year', y='Automobile_Sales', hue='Vehicle_Type')
plt.title("Sales Trend by Vehicle Type: Recession vs Non-Recession", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Automobile Sales", fontsize=12)
plt.legend(title="Vehicle Type / Recession")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
```

4. Now you want to compare the sales of different vehicle types 
during a recession and a non-recession period.


Hint: To visualize sales of different vehicles during recession and non- recession periods, you 
can use a bar chart. You will need to group Recession, Vehicle_Typefor average 
Automobile_Sales and then plot it. Make use of sns. bar plot (x=x, y=data = df))

```python
grouped_avg = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()
plt.figure(figsize=(14, 6))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=grouped_avg)
plt.title("Average Automobile Sales by Vehicle Type\nDuring Recession vs Non-Recession", fontsize=16)
plt.xlabel("Vehicle Type", fontsize=12)
plt.ylabel("Average Sales", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Recession (1=Yes, 0=No)")
plt.tight_layout()
```
